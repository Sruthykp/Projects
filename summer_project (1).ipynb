```python
# Select relevant columns for vaccination data
vaccination_columns = [
    'location', 'date', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated',
    'total_tests', 'new_tests', 'total_tests_per_thousand', 'new_tests_per_thousand',
    'new_tests_smoothed', 'new_tests_smoothed_per_thousand', 'population_density',
    'median_age', 'aged_65_older', 'aged_70_older', 'hospital_beds_per_thousand',
    'handwashing_facilities', 'life_expectancy', 'human_development_index'
]
vaccination_data_asia = data_asia[vaccination_columns]

# Select relevant columns for death data
death_columns = [
    'location', 'date', 'new_cases', 'total_cases', 'new_deaths', 'total_deaths'
]
death_data_asia = death_data_asia[death_columns]
```

```python
# Convert date columns to datetime format
vaccination_data_asia['date'] = pd.to_datetime(vaccination_data_asia['date'])
death_data_asia['date'] = pd.to_datetime(death_data_asia['date'])

# Extract year and month for grouping
vaccination_data_asia['year_month'] = vaccination_data_asia['date'].dt.to_period('M')
death_data_asia['year_month'] = death_data_asia['date'].dt.to_period('M')

# Group by country and month and aggregate the data
monthly_vaccination_data_asia = vaccination_data_asia.groupby(['location', 'year_month']).agg({
    'total_vaccinations': 'sum',
    'people_vaccinated': 'sum',
    'people_fully_vaccinated': 'sum',
    'total_tests': 'sum',
    'new_tests': 'sum',
    'total_tests_per_thousand': 'mean',
    'new_tests_per_thousand': 'mean',
    'new_tests_smoothed': 'sum',
    'new_tests_smoothed_per_thousand': 'mean',
    'population_density': 'mean',
    'median_age': 'mean',
    'aged_65_older': 'mean',
    'aged_70_older': 'mean',
    'hospital_beds_per_thousand': 'mean',
    'handwashing_facilities': 'mean',
    'life_expectancy': 'mean',
    'human_development_index': 'mean'
}).reset_index()

monthly_death_data_asia = death_data_asia.groupby(['location', 'year_month']).agg({
    'new_cases': 'sum',
    'total_cases': 'sum',
    'new_deaths': 'sum',
    'total_deaths': 'sum'
}).reset_index()
```

```python
monthly_data_asia = pd.merge(monthly_vaccination_data_asia, monthly_death_data_asia, on=['location', 'year_month'], how='inner')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error

# Function to cap outliers at a given quantile
def cap_outliers(data, quantile):
    capped_data = data.copy()
    for column in columns_to_check:
        percentile_value = capped_data[column].quantile(quantile)
        capped_data[column] = np.where(capped_data[column] > percentile_value, percentile_value, capped_data[column])
    return capped_data

# Evaluate model performance for different quantiles
for quantile in [0.90, 0.95, 0.99]:
    capped_data = cap_outliers(filtered_data, quantile)
    
    # Split the data
    features = ['total_vaccinations', 'people_fully_vaccinated', 
            'total_tests', 'population_density', 'median_age', 
            'aged_65_older','human_development_index','handwashing_facilities']
    X = capped_data[features]
    y = capped_data['total_deaths'].values.reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=60)
    
    # Feature scaling
    scaler_X = StandardScaler()
    scaler_y = StandardScaler()
    X_train = scaler_X.fit_transform(X_train)
    X_test = scaler_X.transform(X_test)
    y_train = scaler_y.fit_transform(y_train)
    y_test = scaler_y.transform(y_test)
    
    # Create and fit the model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Evaluate the model
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    
    print(f"Quantile {quantile * 100}th - R-squared: {r2}, MSE: {mse}, RMSE: {rmse}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Check if 'year_month' column exists and is of PeriodDtype
if 'year_month' in filtered_data.columns and filtered_data['year_month'].dtype.name == 'period[M]':
    # Convert 'year_month' to datetime format
    filtered_data['year_month'] = filtered_data['year_month'].dt.to_timestamp()

# Set 'year_month' column as the index
filtered_data.set_index('year_month', inplace=True)

# Drop duplicate index entries
filtered_data = filtered_data[~filtered_data.index.duplicated(keep='first')]

# Ensure the DataFrame has a proper frequency
filtered_data = filtered_data.asfreq('M')

# Function to fit and forecast using ARIMA
def fit_arima(train_data, order):
    try:
        model = ARIMA(train_data, order=order)
        model_fit = model.fit()
        return model_fit
    except Exception as e:
        print(f"Error fitting ARIMA model: {e}")
        return None

# Function to evaluate and plot ARIMA model
def evaluate_arima_model(train_data, test_data, arima_order, label):
    arima_model = fit_arima(train_data, arima_order)
    
    if arima_model is None:
        print(f"ARIMA model fitting failed for {label}.")
        return

    # Forecast length must be equal to the length of the test data
    forecast_steps = len(test_data)
    
    # Check if forecast_steps is valid
    if forecast_steps <= 0:
        print(f"Invalid forecast steps for {label}.")
        return

    try:
        pred = arima_model.forecast(steps=forecast_steps)
    except ValueError as e:
        print(f"Error during forecasting for {label}: {e}")
        return

    mse = mean_squared_error(test_data, pred)
    print(f"Mean Squared Error for {label}:", mse)
    
    # Forecast future values
    forecast_period = 5  # Number of months to forecast
    try:
        forecast = arima_model.forecast(steps=forecast_period)
    except ValueError as e:
        print(f"Error during future forecasting for {label}: {e}")
        return
    
    plt.figure(figsize=(14, 7))
    plt.plot(train_data, label='Training Data')
    plt.plot(test_data, label='Test Data')
    plt.plot(test_data.index, pred, label=f'Predicted {label}')
    forecast_index = pd.date_range(start=test_data.index[-1], periods=forecast_period + 1, freq='M')[1:]
    plt.plot(forecast_index, forecast, label=f'Forecasted {label}')
    plt.title(f'ARIMA Model - {label}')
    plt.xlabel('Date')
    plt.ylabel(label)
    plt.legend()
    plt.show()

# ARIMA order (p, d, q)
arima_order = (5, 1, 2)  # Adjust this order based on your data and model diagnostics

# Split the data into training and testing sets
train_end = '2023-04'
test_start = pd.to_datetime(train_end) + pd.DateOffset(months=1)
train_data_new_cases = filtered_data.loc[:train_end, 'new_cases']
test_data_new_cases = filtered_data.loc[test_start:, 'new_cases']
evaluate_arima_model(train_data_new_cases, test_data_new_cases, arima_order, 'New Cases')

# Evaluate and plot ARIMA model for total_cases
train_data_total_cases = filtered_data.loc[:train_end, 'total_cases']
test_data_total_cases = filtered_data.loc[test_start:, 'total_cases']
evaluate_arima_model(train_data_total_cases, test_data_total_cases, arima_order, 'Total Cases')

# Evaluate and plot ARIMA model for new_deaths
train_data_new_deaths = filtered_data.loc[:train_end, 'new_deaths']
test_data_new_deaths = filtered_data.loc[test_start:, 'new_deaths']
evaluate_arima_model(train_data_new_deaths, test_data_new_deaths, arima_order, 'New Deaths')

# Evaluate and plot ARIMA model for total_deaths
train_data_total_deaths = filtered_data.loc[:train_end, 'total_deaths']
test_data_total_deaths = filtered_data.loc[test_start:, 'total_deaths']
evaluate_arima_model(train_data_total_deaths, test_data_total_deaths, arima_order, 'Total Deaths')
```
